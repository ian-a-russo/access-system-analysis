```python
import pandas as pd

url = '../data/acessos.csv'

try:
    dados = pd.read_csv(url, sep=';')
except UnicodeDecodeError:
    dados = pd.read_csv(url, sep=';', encoding='latin1', engine='python')

print("Linhas e colunas:", dados.shape)
display(dados.head())
```

```python
dias         = dados["Dia"]
sucesso_pct  = dados["Proporção de Acessos Bem-Sucedidos (%)"]
erro_pct     = dados["Proporção de Operações com Erros (%)"]
stock_pct    = dados["Produtos em Estoque (%)"]
```

```python
import matplotlib.pyplot as plt

# --- Gráfico 1: Proporção de Acessos Bem-Sucedidos ---
plt.figure(figsize=(8,4))
plt.plot(dias, sucesso_pct, marker='o')
plt.title('Proporção de Acessos Bem-Sucedidos (%)')
plt.xlabel('Dia')
plt.ylabel('Sucesso (%)')
plt.xticks(dias)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Gráfico 2: Proporção de Operações com Erros ---
plt.figure(figsize=(8,4))
plt.plot(dias, erro_pct, marker='o')
plt.title('Proporção de Operações com Erros (%)')
plt.xlabel('Dia')
plt.ylabel('Erro (%)')
plt.xticks(dias)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# --- Gráfico 3: Produtos em Estoque (%) ---
plt.figure(figsize=(8,4))
plt.plot(dias, stock_pct, marker='o')
plt.title('Produtos em Estoque (%)')
plt.xlabel('Dia')
plt.ylabel('Estoque (%)')
plt.xticks(dias)
plt.grid(True)
plt.tight_layout()
plt.show()
```
